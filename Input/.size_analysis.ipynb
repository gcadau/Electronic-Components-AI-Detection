```python
from PIL import Image
import os
import statistics
import matplotlib.pyplot as plt

directory = "dataset"
```

```python
for filename in os.listdir(directory):
    file_path = os.path.join(directory, filename)
    with Image.open(file_path) as img:
        width, height = img.size
        print(f"{filename}: {width}x{height}")
```

```python
dimensions_array = []

for filename in os.listdir(directory):
    file_path = os.path.join(directory, filename)
    with Image.open(file_path) as img:
        width, height = img.size
        dimensions_array.append((width, height))

mean_dimensions = (
    sum(dim[0] for dim in dimensions_array) / len(dimensions_array),
    sum(dim[1] for dim in dimensions_array) / len(dimensions_array),
)

variance_dimensions = (
    statistics.variance(dim[0] for dim in dimensions_array),
    statistics.variance(dim[1] for dim in dimensions_array),
)

std_dev_dimensions = (
    statistics.stdev(dim[0] for dim in dimensions_array),
    statistics.stdev(dim[1] for dim in dimensions_array),
)

median_dimensions = (
    statistics.median(dim[0] for dim in dimensions_array),
    statistics.median(dim[1] for dim in dimensions_array),
)

mode_dimensions = (
    statistics.mode(dim[0] for dim in dimensions_array),
    statistics.mode(dim[1] for dim in dimensions_array),
)

print("Mean dimensions:", mean_dimensions)
print("Variance dimensions:", variance_dimensions)
print("Standard deviation dimensions:", std_dev_dimensions)
print("Median dimensions:", median_dimensions)
print("Mode dimension:", mode_dimensions)
```

```python
widths_array = [dim[0] for dim in dimensions_array]
heights_array = [dim[1] for dim in dimensions_array]


plt.figure(figsize=(20, 12))
plt.hist(widths_array, bins=20, color='blue', edgecolor='black', alpha=0.7, label='Widths')
plt.hist(heights_array, bins=20, color='red', edgecolor='black', alpha=0.7, label='Heights')
plt.title('Distribution of Image Dimensions')
plt.xlabel('Dimensions')
plt.ylabel('Frequency')
plt.legend(loc='upper right')
plt.show()
```

```python
min_width, max_width = min(widths_array), max(widths_array)
min_height, max_height = min(heights_array), max(heights_array)

print(f"Min Width: {min_width}, Max Width: {max_width}")
print(f"Min Height: {min_height}, Max Height: {max_height}")
```
